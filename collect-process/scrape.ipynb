```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromiumService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()
# options.add_argument('--headless=new')

url = "https://dsc-capstone.org/"
driver = webdriver.Chrome(service=ChromiumService(), options=options)
driver.get(url)

wait = WebDriverWait(driver, timeout=20, poll_frequency=1)
el = wait.until(lambda d: d.find_element(By.LINK_TEXT, "About"))
print(el)

driver.quit()

## supposedly you don't need webdriver manager as in selenium 4.6.0 and above, the driver is auto managed
# https://www.selenium.dev/documentation/webdriver/troubleshooting/errors/driver_location/
```
