```python
from pathlib import Path
import pandas as pd
import glob
import os

out_dir = Path("./processed/")
out_dir.mkdir(parents=True, exist_ok=True)

path = "./202501/"
date = "20250101T185049"

matches = glob.glob(os.path.join(path, f"*{date}*"))

p = Path(matches[1])
a = pd.read_csv(p, usecols=['pid', 'Lingulodinium_polyedra'])
a['pid'] = a['pid'].str.rsplit('_', n=1).str[-1].astype(int)
print(a.head())



out_file = out_dir / f"{date}.csv"
a.to_csv(out_file, index=False)

# features = Path(matches[2])

# hdr = Path(matches[0])
# print(hdr.read_text())
```
